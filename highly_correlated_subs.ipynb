```python
#get subtypes with correlation>0.91
high_corr = merged_df[merged_df['median_correlation'] > 0.91]
high_corr=high_corr.drop_duplicates()

def get_lineage(high_corr, tcga_meta):
    """ highcorr is a dataframe that containts the highly correlated subtypes for each annotated_subtype in the top5"""
    data = []
    for subtype in high_corr.top5.unique():
        lineage_values = tcga_meta[tcga_meta.derived_subtype == subtype]['derived_lineage'].unique()
        
        if len(lineage_values) == 1:
            lineage = lineage_values[0]
        else:
            lineage = tcga_meta[tcga_meta.derived_subtype == subtype]['derived_subtype'].unique()
        
        data.append((subtype, lineage))
    
    return pd.DataFrame(data, columns=['top5', 'lineage'])
```

```python
#finds lineage for the top5 subtypes with correlation>0.91 for each annotated_subtype 
subtype_lineage= get_lineage(high_corr, tcga_meta)
subtype_lineage
```

```python
#
def get_samelineage(data):
    """
    Filters out subtypes that have more than one unique lineage from the given DataFrame.

    iterates through unique subtypes in the 'top5' column of the input DataFrame,
    identifies subtypes that are associated with more than one unique lineage in the 'lineage' column,
    and returns a new DataFrame that excludes those subtypes.
    - data (pd.DataFrame): A DataFrame containing the top5 subtypes for each annotated_subtype that have highcorrelation, their score and median_corr
    Returns a Dataframe that includes the lineage for the top5 subtypes for each annotated_subtype
     """
    exclude_subtypes = []
    for subtype in data.top5.unique():
        if len(data[data.top5== subtype]['lineage']) > 1:
            exclude_subtypes.append(subtype)

    filtered_df = data[~data['top5'].isin(exclude_subtypes)]
    
    return filtered_df    
```

```python
#update the oncotree Data 
from taigapy import create_taiga_client_v3
tc = create_taiga_client_v3()
oncotree = tc.get(name="oncotree-42c7", version=1, file="oncotree").set_index("code")
oncotree=oncotree.reset_index()

# Step 1: Find names with more than one unique OncoTree code
multiple_codes = oncotree.groupby('name')['code'].nunique()
names_with_multiple_codes = multiple_codes[multiple_codes > 1].index

# Step 2: Update the 'name' column in the oncotree DataFrame
def update_name(row):
    if row['name'] in names_with_multiple_codes:
        return f"{row['name']} ({row['tissue']})"
    return row['name']

oncotree['name'] = oncotree.apply(update_name, axis=1)
oncotree=oncotree.set_index('code')
def get_sublineage(code,oncotree):
    #takes an oncotree code and gives you back the node on the oncotree thats just below the lineage level
    node = oncotree.loc[code]
    while oncotree.loc[node.parent].tissue != oncotree.loc[node.parent]['name']:
        node = oncotree.loc[node.parent]
    return node['name']
```

```python
# fill out the subtypes with no oncotree code 
glioma_row = oncotree.loc['GNOS']
glioma_row['name'] = 'glioma'
glioma_row.name = 'GNOS_1'
oncotree=pd.concat([oncotree,glioma_row.to_frame().T])
acute_Row=oncotree.loc['TLL']	
acute_Row['name'] = 'acute lymphoblastic leukemia'
acute_Row.name = 'TLL_1'
ocotree=pd.concat([oncotree,acute_Row.to_frame().T])
```

```python
#find the nearest common parent OncoTree code between two cancer subtypes (top5 and annotated_subtype) within the same lineage
def get_common_parent_code(code_1, code_2, oncotree):
    def find_parents(code, oncotree):
        parents = []
        while code in oncotree.index:
            parent_code = oncotree.loc[code, 'parent']
            if pd.isna(parent_code):
                break
            parents.append(parent_code)
            code = parent_code
        return parents

    parents_1 = find_parents(code_1, oncotree)
    parents_2 = find_parents(code_2, oncotree)

    for parent in parents_1:
        if parent in parents_2:
            return parent

    return None
    
#get the nearest common parent between codes for top5subtypes of the same lineage for each annotated_subtype
def find_common_parent(row, tcga_meta, oncotree):
    code_1 = tcga_meta.loc[tcga_meta['derived_subtype'] == row['top5'], 'oncotree_code'].values[0]
    code_2 = tcga_meta.loc[tcga_meta['derived_subtype'] == row['annotated_subtype'], 'oncotree_code'].values[0]
    return get_common_parent_code(code_1, code_2, oncotree)
    
filtered=result_df[['top5', 'annotated_subtype']] 

filtered['common_parent'] = filtered.apply(lambda row: find_common_parent(row, tcga_meta, oncotree), axis=1)
```

```python
#filtered=result_df[['top5', 'annotated_subtype']] 
def get_level(filtered, oncotree):
    """
    Merges the filtered DataFrame with the oncotree DataFrame to get the level of each common parent.

    This function merges the 'filtered' DataFrame with the 'oncotree' DataFrame based on the 'common_parent' column,
    obtaining the hierarchical level of each common parent in the oncotree hierarchy.

    Parameters:
    - filtered (pd.DataFrame): DataFrame containing 'top5' and 'common_parent' columns.
    - oncotree (pd.DataFrame)

    Returns:
    - pd.DataFrame: A merged DataFrame with the 'level' of each 'common_parent' included.
"""
    merged_df = filtered.merge(oncotree[['parent', 'level']], left_on='common_parent', right_on='parent', how='left')
    merged_df = merged_df.rename(columns={'level_y': 'level'})
    
    return merged_df
```

```python
def map_subtype(row, df):
    """
    Maps the 'derived_subtype' in the given row to a new the subtype based on the common_parent for the subtypes 
    in that have high correlation and same lineage for each annotated subtype

    Parameters:
    - row (pd.Series): A row of the DataFrame being processed.
    - df (pd.DataFrame): The DataFrame containing the top5,	annotated_subtype,	parent,	level and the new_subtype column which comes from the oncotree data 

    Returns:
    - str: The new subtype if a match is found; otherwise, returns the original 'derived_subtype'.
    """
    # Check if 'derived_subtype' in the row exists in the 'top5' column of df
    if row['derived_subtype'] in df['top5'].values:
        # If it exists, return the corresponding 'new_subtype' value
        return df[df['top5'] == row['derived_subtype']]['new_subtype'].values[0]
    # Check if 'derived_subtype' in the row exists in the 'annotated_subtype' column of df
    elif row['derived_subtype'] in df['annotated_subtype'].values:
        # If it exists, return the corresponding 'new_subtype' value
        return df[df['annotated_subtype'] == row['derived_subtype']]['new_subtype'].values[0]
    else:
        # If no match is found, return the original 'derived_subtype' value
        return row['derived_subtype']

# Apply the map_subtype function to each row of the tcga_meta DataFrame
tcga_meta['new_subtype'] = tcga_meta.apply(lambda row: map_subtype(row, df), axis=1)
```

```python
def update_new_subtype_cns(df):
    """Maps subtypes with a lineage of CNS/BRAIN to 'CNS/BRAIN', except for subtypes that contain 'Medulloblastoma'.
    Other subtypes are mapped to their original subtype value."""
    def update_subtype(row):
        if row['derived_lineage'] == 'CNS/Brain':
            if 'medulloblastoma' not in row['subtype'].lower():
                return 'CNS/Brain'
        return row['new_subtype']

    # Apply the update_subtype function to each row
    df['new_subtype'] = df.apply(update_subtype, axis=1)
    return df 
tcga_meta=update_new_subtype_cns(tcga_meta)
```
